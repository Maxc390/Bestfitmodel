```python
import numpy as np
import csv
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import curve_fit
from numpy import cos
```

```python
arr=(np.loadtxt("set.txt",dtype='float64',usecols=(0,1),unpack=True))
```

```python
arr.shape
```

```python
Temp=arr[0,:]
```

```python
Date=arr[1,:]
```

```python
Date = np.array(Date)
Temp = np.array(Temp)
t=Date
```

```python
#f(t) = a cost(2pi*t + b) +c
# Define the function to fit
def equation(t,a, b, c):
    return a*cos(t)*(2.0*np.pi*t+ b)+ c
```

```python
guess=[100,100,20]
guess=np.asarray(guess)
```

```python
# Perform the curve fit
popt,pcov = curve_fit(equation, Date, Temp,guess)
print(popt)
print(pcov)
```

```python
#optimum values
a_opt,b_opt,c_opt = popt
x_model= np.arange(min(Date),max(Date),a_opt)
y_model=equation(x_model,a_opt,b_opt*5,c_opt)
plt.xlabel("Years")
plt.ylabel("Temperature")

plt.plot(Date,Temp,'o')
plt.plot(x_model, y_model, 'r-', label='Fit')
plt.show()
```
